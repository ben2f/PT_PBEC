```python
from numpy import *
from matplotlib import *
from pylab import *
import pandas as pd
from tqdm import tqdm_notebook as tqdm
```

```python
df = pd.read_csv("real1d-dyn.txt" , sep='\\s+', header=None)
```

```python
df.columns = ['x', 't', 'z']
x = df['x'].values
t = df['t'].values
u = df['z'].values
```

```python
# Step 2: Reshape u to 2D array (assuming you know the shape from x and t)
n_x = len(unique(x))
n_t = len(unique(t))
u_reshaped = u.reshape((n_t, n_x))  # Reshaping u

# Step 3: Plot using imshow (flipud to match flipud in MATLAB)
figure()
figure(figsize=(3,6))
imshow(flipud(u_reshaped), aspect='auto', extent=[x.min(), x.max(), t.min(), t.max()],cmap='inferno')
#cmap='jet'
#cmap='inferno'
colorbar(label='$|\\psi|^2$')
xlabel('x')
ylabel('t')
#xlim(-12,12)
#ylim(0,25)
#title("g=5,$g_1=0.01$,$\\gamma=-0.001$,$NH =0.01$")
title("g=-10,W=0.01")
savefig("7b_0.01.png",dpi=1200)
#savefig("5b_0.1.jpg",dpi=1200)
show()
```
