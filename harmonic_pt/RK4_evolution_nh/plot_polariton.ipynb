```python
from numpy import *
from matplotlib import *
from pylab import *
import pandas as pd
```

```python
df1 = pd.read_csv("LS0.01.txt" , sep='\\s+', header=None)
df2 = pd.read_csv("LS0.5.txt" , sep='\\s+', header=None)
```

```python
t1=df1[0]
x1=df1[1]
y1=df1[2]
```

```python
t2=df2[0]
x2=df2[1]
y2=df2[2]
```

```python
fig, ax1 = subplots()

ax1.plot(t1,x1,linestyle='-', linewidth=0.7, color='blue',label="$\\gamma_{eff}=0.01) $")
ax1.set_xlabel('Time')
ax1.set_ylabel('amplitude ($\\gamma_{eff}=0.01) $')
ax1.set_xlim(0,120)
ax1.set_ylim(0,40)
ax1.tick_params(axis='y', labelcolor="blue")
title('"g=5,$g_1=0.01$,,$V_0 =0.03125$"')
legend()

ax2 = ax1.twinx()  
ax2.set_ylabel('amplitude ($\\gamma_{eff}=0.5) $')
ax2.plot(t2,x2,linestyle='-', linewidth=0.7, color='red',label="$\\gamma_{eff}=0.5) $")
ax2.set_ylim(0,40)
ax2.tick_params(axis='y', labelcolor='red')
legend()
fig.tight_layout() 
savefig("AT_b.png",dpi=3000, bbox_inches='tight')
show()
```

```python
fig, ax1 = subplots()

ax1.plot(t1,y1,linestyle='-', linewidth=0.7, color='blue',label="$\\gamma_{eff}=0.01) $")
ax1.set_xlabel('Time')
ax1.set_ylabel('Width ($\\gamma_{eff}=0.01) $')
ax1.set_xlim(0,120)
ax1.set_ylim(0,250)
ax1.tick_params(axis='y', labelcolor="blue")
title('"g=5,$g_1=0.01$,,$V_0 = 0.03125$"')
#legend()

ax2 = ax1.twinx()  
ax2.set_ylabel('Width ($\\gamma_{eff}=0.5) $')
ax2.plot(t2,y2,linestyle='-', linewidth=0.7, color='red',label="$\\gamma_{eff}=0.5) $")
ax2.set_ylim(0,250)
ax2.tick_params(axis='y', labelcolor='red')
#legend()
fig.tight_layout() 
savefig("RT_b.png",dpi=3000, bbox_inches='tight')
show()
```
